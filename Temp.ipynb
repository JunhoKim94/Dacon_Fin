```python
import konlpy
import pandas as pd
import numpy as np
from konlpy.tag import *
from collections import Counter
```

```python
xx = pd.read_csv("./data/train.csv")
```

```python
batch_size = 100
seed = np.random.choice(len(xx), 100)
```

```python
stopwords = ['XXX', '.', '을', '를', '이', '가', '-', '(', ')', ':', '!', '?', ')-', '.-', 'ㅡ', 'XXXXXX', '..', '.(', '은', '는'] #필요없는 단어 리스트
```

```python
#tokenize
raw_x = xx.iloc[:,2]
raw_y = xx.iloc[:,-1]
```

```python
batch_x = raw_x.iloc[seed]
batch_y = raw_y.iloc[seed]
```

```python
komoran = Komoran()
```

```python
kkma = Kkma()
```

```python
def tokenizing(batch_size ,tokenizer = Kkma()):
    
    tokenizer = tokenizer
    elements = dict()

    for iteration in range(batch_size):
        temp = tokenizer.morphs(batch_x.iloc[iteration])
        for item in temp:
            if item not in elements.keys():
                num = len(elements)
                elements[item] = num
                
    return elements
```
